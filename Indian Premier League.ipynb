### Imporing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Imporing the Dataset

```python
df=pd.read_csv('matches.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.drop('umpire3',axis=1) # there are more null values present in feature 'umpire 3', So dropping this feature.. 
```

```python
df.describe()
```

```python
print("Number of Seasons played:", len(df['season'].unique()))
```

```python
print("No. of Matches Played")
df['season'].value_counts()
```

### Visualization

```python
plt.figure(figsize=(15,8))
sns.countplot(x='season', data=df)
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x='winner', data=df)
plt.title('Match Winning Teams')
plt.xticks(rotation=90)
```

```python
plt.subplots(figsize=(15,8))
sns.countplot(x ='venue',data=df,order=df['venue'].value_counts().index)
plt.title('Indian Premier League Venue',fontsize=35)
plt.xlabel('Stadium',fontsize=15)
plt.ylabel('Matches',fontsize=15)
plt.xticks(rotation=-90,fontsize=15)
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x= df.player_of_match.value_counts()[:10].index,y= df.player_of_match.value_counts()[:10])
plt.title('Top player of the match Winners',fontsize=35)
plt.xlabel('Player',fontsize=25)
plt.ylabel('Match',fontsize=25)
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.countplot('winner', data = df)
plt.title('Champion Teams',fontsize=35)
plt.xlabel('Teams',fontsize=15)
plt.ylabel('Season won',fontsize=15)
plt.xticks(rotation=-90,fontsize=15)
plt.show()
```

### Conclusion

Mumbai Indians plays the highest amount of matches and has won more seasons than the other teams
Chennai Super Kings ranked the second highest in winning matches followed by Kolkata Knight Riders
Eden Gardens hosts the highest number of matches followed by M Chinnaswamy Stadium and Wankhede Stadium
CH Gayle won the most MVP awards in Indian Premier League
